## **1. Introduction**

For Project 1 of SPDB 2025/2026, the analysis will focus on the following research questions:

**1. Do taxi rides exhibit geographic and/or temporal patterns?**

**2. Which times of day are more profitable for taxi rides?**

In 1% of New York City taxi trip data, the main objectives are to determine whether there are trends in the most frequently requested locations for these trips, identify patterns associated with different times of day, and investigate which times of day are associated with higher profitability for taxi rides.

By using graphs and visualizations, this analysis combines location data (pickup/dropoff grids) with temporal data (hour, day of the week, month) and revenue data to provide a comprehensive view of taxi trip dynamics and profitability trends.

## **3. Methodology: Do taxi rides exhibit geographic and/or temporal patterns?**

### **3.1 Spacial Analysis**

The spatial analysis of NYC taxi trips was conducted following a systematic workflow to identify hotspots of pickup and dropoff activity and the most frequent routes.

**AI promps:**

1. ChatGPT was provided with all the contents of the SPBD2526_Proj1.ipynb notebook. "Focusing on the topic "Do taxi rides exhibit geographic and/or temporal patterns?". The aim is to perform a Spatial Pattern Analysis, based on the most frequent areas, please provide me the code that allows me to perform this task."
- The code provided by ChatGPT covered several topics, including the most frequent zones, frequent zones represented in heatmaps, and frequent zones represented as hotspots.
- In many cases, the code examples focused only on pickups; however, these were easily adapted manually to handle dropoffs as well.
- The comments provided directly in the code by ChatGPT were also manually translated into English to maintain clarity and consistency.

2. The code for "Most Frequency Routes" was provided. "I wanted a graphical visualization method for better analysis of these routes."
- Initially, the code encountered execution issues. The exact error messages were copied and provided to ChatGPT, which then generated alternative solutions. After several attempts, it was possible to successfully generate heatmaps highlighting both the most and least frequent routes.

**Data Preparation:**

The dataset was first filtered to remove records with missing values in critical geographical (pickup_latitude, pickup_longitude, dropoff_latitude, dropoff_longitude) and temporal (pickup_datetime, dropoff_datetime) columns.
Temporal features were extracted from the pickup and dropoff timestamps to allow for further temporal analyses.
Spatial Grid Mapping:
The geographic area of New York City was divided into a uniform grid based on latitude and longitude ranges.
Each pickup and dropoff location was mapped to a corresponding grid cell.
Trips outside the defined grid boundaries were excluded to focus the analysis on valid city areas.

**Aggregation of Pickup and Dropoff Events:**

The number of pickups and dropoffs was aggregated per grid cell.
To identify high-traffic areas, only the top 5% of grid cells by trip count (95th percentile) were considered as hotspots.
Routes were defined as pairs of pickup and dropoff grid cells.
The frequency of trips along each route was calculated, allowing the identification of the most commonly traveled paths in the city.

**Visualization:**

Heatmaps were generated for both pickups and dropoffs using a scatter plot representation of the grid.
The size and color of each point were scaled according to the number of trips in the corresponding cell, providing an intuitive visual identification of hotspots.

Heatmaps were generated for the most frequent taxi routes, representing trips from pickup cells to dropoff cells. The size and color of each point were scaled according to the number of trips for the corresponding route, providing an intuitive visual representation of the busiest and least traveled routes in the city. Initial execution issues occurred, but these were resolved by using the error messages to generate alternative solutions, allowing the heatmaps to be successfully produced.

**Data Preparation:**

Using these features, the total number of rides was aggregated and examined by:

- Day of the week, to observe weekly patterns in ride activity.

This analysis provides insights into how ride activity varies over time, helping to identify trends and patterns in user behavior.

## **4. Methodology: Which times of day are more profitable for taxi rides?**

### **4.1 Temporal Analysis of Profitability**

The temporal analysis of NYC taxi trips was conducted following a systematic workflow to identify the hours of the day, days of the week, and months associated with higher revenue and profitability

**AI promps:**
1. ChatGPT was provided with all the contents of the SPBD2526_Proj1.ipynb notebook. "Focusing on the topic "Which times of day are more profitable for taxi rides?". I wanted to analyze more specific temporal patterns related to revenue generation, such as profitability based on the hour of the day, the day of the week, and the month of the year."
- The code provided by ChatGPT was mostly for pickup data, and it was manually adjusted to also handle dropoff data.

**Data Preparation:**

For this part of the project, the dataset had already been cleaned, filtered, and prepared, so no additional data preprocessing was required prior to the analysis.

**Hour of the day:**

The dataset was aggregated by pickup and dropoff hours. The average revenue per trip (total_amount) was calculated for each hour to identify which times of day are more profitable.

**Day of the week:**

Trips were grouped by pickup and dropoff days of the week. The average revenue per trip was computed for each day, allowing the detection of weekday vs. weekend profitability patterns.

**Month of the year:**

Data was aggregated by pickup and dropoff months. Average revenue per trip was analyzed to explore potential seasonal trends in profitability.
